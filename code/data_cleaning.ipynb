```python
unclean_crime = unclean_crime[['Case Number', 'Date', 'Primary Type', 'Latitude', 'Longitude']]
unclean_crime
```

```python
unclean_crime = unclean_crime[unclean_crime.Latitude.isnull() == False]
unclean_crime
```

```python
unclean_crime.reset_index(drop=True, inplace=True)
```

```python
unclean_crime = unclean_crime.sort_values(by='Date').reset_index(drop=True)
```

```python
unclean_crime = unclean_crime.rename(columns={'Case Number' : 'id', 'Date': 'date', 'Primary Type' : 'type', 'Latitude' : 'lat', 'Longitude' : 'Long'})
unclean_crime
```
