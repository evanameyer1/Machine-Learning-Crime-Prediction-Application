```python
import pandas as pd
import datetime
```

```python
unclean_alleylights = pd.read_csv('../original_datasets/unclean_alleylights.csv')
unclean_alleylights
```

```python
def clean_data(df):
    df = df[df.Status == 'Completed']
    df = df[['Creation Date', 'Completion Date', 'Service Request Number', 'Latitude', 'Longitude']]
    df.rename(columns={'Creation Date' : 'start_date', 'Service Request Number' : 'id', 
                       'Latitude': 'lat', 'Longitude' : 'long', 'Completion Date' : 'end_date'}, inplace=True)
    df.dropna(subset=['lat', 'long'], inplace=True)
    df['start_date'] = pd.to_datetime(df['start_date'], format='%m/%d/%Y', errors='coerce')
    df['end_date'] = pd.to_datetime(df['end_date'], format='%m/%d/%Y', errors='coerce')
    df['duration'] = df.end_date - df.start_date
    df.sort_values(by='start_date', inplace=True)
    df.reset_index(drop=True, inplace=True)
    return df
```

```python
unclean_alleylights = clean_data(unclean_alleylights).tail(-4)
unclean_alleylights
```

```python
unclean_streetlights_allout = pd.read_csv('../original_datasets/unclean_streetlights_allout.csv')
unclean_streetlights_allout
```

```python
unclean_streetlights_allout = clean_data(unclean_streetlights_allout)
unclean_streetlights_allout
```

```python
unclean_streetlights_oneout = pd.read_csv('../original_datasets/unclean_streetlights_oneout.csv')
unclean_streetlights_oneout
```

```python
unclean_streetlights_oneout = clean_data(unclean_streetlights_oneout)
unclean_streetlights_oneout
```

```python
clean_lighting = pd.concat([unclean_alleylights, unclean_streetlights_allout, unclean_streetlights_oneout])
clean_lighting
```

```python
clean_lighting.sort_values(by='start_date', inplace=True)
clean_lighting.reset_index(drop=True, inplace=True)
clean_lighting
```

```python
data = []
for i in range(len(clean_lighting)):
    duration = clean_lighting.duration.iloc[i].days + 1
    for j in range(duration):
        date = clean_lighting.start_date.iloc[i] + datetime.timedelta(days=j)
        id = clean_lighting.id.iloc[i]
        lat = clean_lighting.lat.iloc[i]
        long = clean_lighting.long.iloc[i]
        d = {'date' : date, 'id' : id, 'lat' : lat, 'long' : long}
        data.append(d)
temp = pd.DataFrame(data)
```

```python
clean_lighting = temp
clean_lighting
```

```python
clean_lighting.to_csv('../clean_datasets/clean_lighting.csv')
```

```python
unclean_vacant_buildings = pd.read_csv('../original_datasets/unclean_vacant_buildings.csv')
unclean_vacant_buildings
```

```python
unclean_vacant_buildings = unclean_vacant_buildings[['DATE SERVICE REQUEST WAS RECEIVED', 'SERVICE REQUEST NUMBER', 'LATITUDE', 'LONGITUDE']]
unclean_vacant_buildings.rename(columns={'DATE SERVICE REQUEST WAS RECEIVED' : 'start_date', 'SERVICE REQUEST NUMBER' : 'id', 'LATITUDE': 'lat', 'LONGITUDE' : 'long'}, inplace=True)
unclean_vacant_buildings['start_date'] = pd.to_datetime(unclean_vacant_buildings['start_date'], format='%m/%d/%Y', errors='coerce')
unclean_vacant_buildings['end_date'] = datetime.datetime(2023, 10, 19)
unclean_vacant_buildings['duration'] = unclean_vacant_buildings['end_date'] - unclean_vacant_buildings['start_date']
unclean_vacant_buildings
```

```python
clean_vacant_buildings = pd.read_csv('../clean_datasets/clean_vacant_buildings.csv')
clean_vacant_buildings = clean_vacant_buildings[['start_date', 'id', 'lat',
       'long', 'end_date', 'duration', 'status', 'area']]
clean_vacant_buildings
```

```python
clean_vacant_buildings.start_date = pd.to_datetime(clean_vacant_buildings.start_date)

data = []
for i in range(len(clean_vacant_buildings)):
    duration = int(clean_vacant_buildings.duration.iloc[i].split(' ')[0]) + 1
    id = clean_vacant_buildings.id.iloc[i]
    lat = clean_vacant_buildings.lat.iloc[i]
    long = clean_vacant_buildings.long.iloc[i]
    status = clean_vacant_buildings.status.iloc[i]
    area = clean_vacant_buildings.area.iloc[i]
    for j in range(duration):
        date = clean_vacant_buildings.start_date.iloc[i] + datetime.timedelta(days=j)
        d = {'date' : date, 'id' : id, 'lat' : lat, 'long' : long, 'status' : status, 'area' : area}
        data.append(d)
temp = pd.DataFrame(data)
```

```python
unclean_vacant_buildings = temp
```
