```python
def read_in_dfs(tables):
    df_list = []
    for table in tables:
        globals()[table.split('.')[0]] = pd.read_csv(f"../datasets/clean/{table}")
        df_list.append(globals()[table.split('.')[0]])
        print(f'{table.split(".")[0]} is done')
```

```python
exempt = ['clean_ridership', 'clean_ridership_totals', 'clean_divvy_trips', 'clean_disadvantaged_areas', 'clean_areas', 'clean_public_healthindicator']
```

```python
def gather_files(directory):
    tables = []
    for foldername, subfolders, filenames in os.walk(directory):
        for filename in filenames:
            if filename.split('.')[0] not in exempt:
                tables.append(filename)
                print(f'{filename} successfully identified')
    return tables
```

```python
root_directory = '../datasets/clean/'
read_in_dfs(gather_files(root_directory))
```
