```python
import pandas as pd
```

```python
# load the dataset
df = pd.read_csv('../datasets/scoring/final_aggregation.csv')
```

```python
df['hour'] = df['hour'].apply(lambda x: '00:00' if x == 0 else f'{x - 1:02}:00')

# Combine 'date' and 'hour' columns into a single 'timestamp' column
df['timestamp'] = df['date'] + ' ' + df['hour']

# Convert 'timestamp' column to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'], format='%Y-%m-%d %H:%M')
```

```python
df
```

```python
df['date'] = df['timestamp']
df.drop(columns=['hour', 'timestamp'], inplace=True)
df.head()
```

```python
df.dtypes
```

```python
def split_table():
    df_list = {}
    for i in range(1, df.area.nunique()):

        table1_name = 'final_' + str(i)
        table2_name = 'ml_' + str(i)

        globals()[table1_name] = df[df.area == i]
        globals()[table1_name].drop(columns='area', inplace=True)
        globals()[table1_name].sort_values(by='date', inplace=True)
        globals()[table1_name].reset_index(drop=True, inplace=True)

        globals()[table2_name] = globals()[table1_name][['date', 'non-violent', 'violent', 'train_rides', 'bike_rides', 'lighting', 'vacant_buildings']]

        df_list[table1_name] = globals()[table1_name]
        df_list[table2_name] = globals()[table2_name]

        df_list[table2_name].to_csv(f'../datasets/training/SARIMA/by_area/{table2_name}.csv', index=False)
        df_list[table1_name].to_csv(f'../datasets/scoring/by_area/{table1_name}.csv', index=False)
```
